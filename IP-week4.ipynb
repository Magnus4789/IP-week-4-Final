**Data Analysis**



```python
#Identify the most popular hour of the day for picking up a shared electric
# car (Bluecar) in the city of Paris over the month of April 2018.
#sorting out cities to remain with paris  only
#assigning paris table as df_p
df_p=df.loc[df['City'] == 'Paris']
df_p
```

```python
#dropping unnecessary columns in df_p

df_p=df.drop(['Address', 'Charging Status', 
       'Displayed comment', 'ID', 'Kind', 'Geo point', 'Scheduled at', 
       'Status', 'Subscription status', 'year', 'month'], axis = 1)
df_p
```

```python
#confirming dropped column
df_p.columns
```

```python
#assigning new variable df_pblue to denote Bluecar counter in Paris and also  adding Utilib & Utilib 1.4 counters
#arranging in ascendin order to Identify the most popular hour of the day for picking up a shared electric car
# (Bluecar) in the city of Paris over the month of April 2018.
#This would mean that all cars should be present at the station that is 7
df_p=df_p.sort_values( by="Bluecar counter",   ascending=False)
df_p.head(7)
```

the most popular hour of the day for picking up a shared electric car(Bluecar) in the city of Paris over the month of April 2018 was on hour 11.

```python
#What is the most popular hour for returning cars?
df_pblue =df_p.sort_values( by="Slots",   ascending=False)
df_pblue.head(7)
```

The most popular hour of returning would be 14th hour in Yerres as the number of slots are full.

```python
#What station is the most popular?
#here we are looking at station name and ranking it by availability
df_pblue =df_p.sort_values( by=["Slots"],
                           ascending=True)
df_pblue.head(7)
```

The Paris/VouillÃ©/47	favourite station as all the available cars are never present at any given time




```python
#What station had the popular picking hour?
df_pblue =df_p.sort_values( by="Postal code",   ascending=False)
df_pblue.head(7)
```

This would be the  station in Yerres as it is still the most avilable pickup on the 11th hour.

```python
#What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?
#Overall?
#At the most popular picking hour?
df_pblue =df_p.sort_values( by="Bluecar counter",   ascending=False)
df_pblue.head(7)
```

The postal code  the most popular for picking up Blue cars is 93700 of Drancy/Sint Stenay/29 the most popular station (Yerres) however doesnt  belong to that postal code


```python
#Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars?
#1.  most popular hour of the day for picking up a shared e Utilib counterin the city of Paris over the month of April 2018.
df_p=df_p.sort_values( by="Utilib counter",   ascending=False)
df_p.head(7)
```

```python
#Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars?
#2.  most popular hour of the day for picking up a shared e Utilib counterin the city of Paris over the month of April 2018.
df_p=df_p.sort_values( by="Utilib 1.4 counter",   ascending=False)
df_p.head(7)
```
